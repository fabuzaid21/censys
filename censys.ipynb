```python
def extract_cipher_suites(fname, key='domain', attr_name_key='p443.https.tls.cipher_suite.name',
                          attr_name='cipher_suite', num_lines=75000):
    values_dict = {key: [], attr_name: []}
    attr_name_nested_keys = attr_name_key.split('.')
    with open(fname) as f:
        print fname
        i = 0
        for line in f:
            if i % 1000 == 0:
                print i ,
            if i == num_lines:
                break
            obj = json.loads(line)
            key_val = obj[key]
            values_dict[key].append(key_val)
            val = obj
            try:
                for _key in attr_name_nested_keys:
                    val = val[_key]
#                 val = obj['p443']['https']['tls']['cipher_suite']['name']
            except KeyError:
                val = 'N/A'

            values_dict[attr_name].append(val)
            i += 1
    return values_dict
```
